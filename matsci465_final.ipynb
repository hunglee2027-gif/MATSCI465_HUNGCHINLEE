```python
import hyperspy.api as hsa

# loading tif file
s = hsa.load("example_EM_Image.tif")

# 
s.plot()
```

```python
img = (
    s.data['R'].astype(np.float32) +
    s.data['G'].astype(np.float32) +
    s.data['B'].astype(np.float32)
) / 3.0

mean_intensity = img.mean()
std_intensity = img.std()

print(f"Mean intensity: {mean_intensity:.3f}")
print(f"Standard deviation: {std_intensity:.3f}")
```

```python
plt.figure()
plt.imshow(img, cmap="gray")
plt.colorbar(label="Intensity")
plt.title("Grayscale Image (from RGB)")
plt.show()

plt.figure()
plt.hist(img.ravel(), bins=256)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.title("Histogram of Pixel Intensities")
plt.show()
```

### Image statistics and intensity distribution

The mean intensity represents the average signal level of the image, while the standard deviation reflects the contrast and overall variation in pixel intensities.

From the histogram, we observe that the pixel intensities are not uniformly distributed. Most pixels cluster within a limited intensity range, indicating a dominant background signal, while the tail of the distribution corresponds to brighter regions, likely arising from stronger scattering or thicker sample regions.
